## Get KoFam sequences
### nifH (K02588 in KEGG), nifD (K02586), and nifK (K02591)

```python
ko_dict = {'nifH':'K02588', 'nifD':'K02586', 'nifK':'K02591'}
```

```python
# make a file fasta file for each konumber to record fasta  sequences
for ko in ko_dict:
    with open('ko_{}_known.fasta'.format(ko), 'w') as f:
        f.write('')
```

```python
# pull "known" nifHDK seq from Dixon et al
known_nif = pd.read_csv('Nitrogenase_Seqs.csv')

known_nifH = known_nif.NifH.dropna().tolist()
known_nifD = known_nif.NifD.dropna().tolist()
known_nifK = known_nif.NifK.dropna().tolist()

known_list = [known_nifH, known_nifD, known_nifK]
```

```python
for ko, known in zip(ko_dict, known_list):
    # Get ko number
    ko_number = ko_dict[ko]

    # URL to fetch the list of genes associated with the KO number
    url = f"http://rest.kegg.jp/link/genes/{ko_number}"

    # Fetch the list of genes
    response = requests.get(url)
    genes = response.text.strip().split("\n")
    genes = [gene.split("\t")[1] for gene in genes] # only keep gene name (KEGG locus tag)

    # Only keep known genes (Dixon et al)
    genes = [gene for gene in genes if gene.split(":")[1] in known]

    # Saves genes in .fasta file
    for gene in genes:

        # fetch the sequence
        sequence_url = f"http://rest.kegg.jp/get/{gene}/aaseq"
        sequence_response = requests.get(sequence_url)

        # Write the sequence to the file
        with open('ko_{}_known.fasta'.format(ko), 'a') as f:
            f.write(sequence_response.text)

# takes about 2 min (or 20 min without filtering out unknown genes)
```
